# 2 Clusters

```python
import numpy as np
import matplotlib.pyplot as plt
from simpful import *

# Define the fuzzy system
FS = FuzzySystem(show_banner=False)

# Add rules to the fuzzy system
RULE1 = "IF (voltage IS cluster1) THEN (OUTPUT IS fun1)"
RULE2 = "IF (voltage IS cluster2) THEN (OUTPUT IS fun2)"
FS.add_rules([RULE1, RULE2])

# Set output functions for the fuzzy system
FS.set_output_function('fun1', '4.693798+0.046102*voltage')
FS.set_output_function('fun2', '4.734113+-0.009758*voltage')

# Define the fuzzy sets using Gaussian membership functions
FS_1 = FuzzySet(function=Gaussian_MF(6.985386, 2.028348), term='cluster1')
FS_2 = FuzzySet(function=Gaussian_MF(2.793210, 2.141702), term='cluster2')

# Define the linguistic variable 'voltage'
MF_voltage = LinguisticVariable([FS_1, FS_2], concept='voltage')
FS.add_linguistic_variable('voltage', MF_voltage)
```

```python
# Create an array of voltage values for plotting
voltage_values = np.linspace(3, 7, 100)

# Evaluate the output for each voltage value
output_values = []
for v in voltage_values:
    FS.set_variable('voltage', v)  # Set the 'voltage' input value
    result = FS.Sugeno_inference()  # Perform inference
    output_values.append(result['OUTPUT'])  # Extract the output value

# Linear functions
def fun1(voltage):
    return 4.693798+0.046102*voltage

def fun2(voltage):
    return 4.734113+-0.009758*voltage

# Gaussian membership functions
def gaussian(x, mean, std):
    return norm.pdf(x, mean, std)

# Gaussian parameters
mean1, std1 = 6.985386, 2.028348  # Cluster 1
mean2, std2 = 2.793210, 2.141702  # Cluster 2

# Plotting
fig, ax1 = plt.subplots()

# Plot the linear functions
ax1.plot(voltage_values, output_values, label="Fuzzy System Output")
ax1.plot(voltage_values, fun1(voltage_values), label="fun1", color="blue")
ax1.plot(voltage_values, fun2(voltage_values), label="fun2", color="red")
ax1.set_xlabel('Voltage')
ax1.set_ylabel('Output')
#ax1.legend(loc="upper right")

# Create a second y-axis for Gaussian functions
ax2 = ax1.twinx()

# Plot the Gaussian membership functions
ax2.plot(voltage_values, gaussian(voltage_values, mean1, std1), label='Cluster 1 (Gaussian)', color='green', linestyle='--')
ax2.plot(voltage_values, gaussian(voltage_values, mean2, std2), label='Cluster 2 (Gaussian)', color='orange', linestyle='--')
ax2.set_ylabel('Membership Degree')

# Adding legends and titles
fig.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.title('Linear Functions and Gaussian Membership Functions')


plt.show()
```

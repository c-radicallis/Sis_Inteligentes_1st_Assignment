```python
import numpy as np
import matplotlib.pyplot as plt
from simpful import *

# Define the fuzzy system
FS = FuzzySystem(show_banner=False)

# Add rules to the fuzzy system
RULE1 = "IF (voltage IS cluster1) THEN (OUTPUT IS fun1)"
RULE2 = "IF (voltage IS cluster2) THEN (OUTPUT IS fun2)"
FS.add_rules([RULE1, RULE2])

# Set output functions for the fuzzy system
FS.set_output_function('fun1', '4.693798+0.046102*voltage')
FS.set_output_function('fun2', '4.734113+-0.009758*voltage')

# Define the fuzzy sets using Gaussian membership functions
FS_1 = FuzzySet(function=Gaussian_MF(6.985386, 2.028348), term='cluster1')
FS_2 = FuzzySet(function=Gaussian_MF(2.793210, 2.141702), term='cluster2')

# Define the linguistic variable 'voltage'
MF_voltage = LinguisticVariable([FS_1, FS_2], concept='voltage')
FS.add_linguistic_variable('voltage', MF_voltage)

# Create an array of voltage values for plotting
voltage_values = np.linspace(3, 7, 100)

# Evaluate the output for each voltage value
output_values = []
for v in voltage_values:
    FS.set_variable('voltage', v)  # Set the 'voltage' input value
    result = FS.Sugeno_inference()  # Perform inference
    output_values.append(result['OUTPUT'])  # Extract the output value

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(voltage_values, output_values, label="Fuzzy System Output")
plt.xlabel("Input Voltage")
plt.ylabel("Temperature (Voltage)")
plt.title("System Output as a Function of Input Voltage")
plt.grid(True)
#plt.legend()
plt.show()
```
